# Summer Olympics

```python
import pandas as pd
```

```python
df = pd.read_excel('Summer Olympic medallists 1896 to 2008.xlsx',
                   sheet_name='BREAKDOWN', skiprows=[0,1])
```

```python
df.head()
```

```python
editions = df[['Edition', 'Grand Total', 'City', 'Country']]
editions.head()
```

```python
ioc_codes = pd.read_excel('Summer Olympic medallists 1896 to 2008.xlsx',
                       sheet_name='IOC COUNTRY CODES')
```

```python
ioc_codes = ioc_codes[['Country','ISO code']]
ioc_codes.head()
```

```python
medals = pd.read_excel('Summer Olympic medallists 1896 to 2008.xlsx',
                       sheet_name='ALL MEDALISTS', skiprows=[0,1,2,3])
medals = medals[['Edition', 'Athlete', 'NOC', 'Medal']]
medals.head()
```

```python
medal_counts = medals.pivot_table( index= 'Edition', values= 'Athlete', columns= 'NOC', aggfunc = 'count')
medal_counts.head()
```

```python
# Left join editions and ioc_codes: hosts
hosts = pd.merge(editions,ioc_codes, how='left')

# Extract relevant columns and set index: hosts
hosts = hosts[['Edition','ISO code']].set_index('Edition')

# Fix missing 'NOC' values of hosts
print(hosts.loc[hosts['ISO code'].isnull()])
hosts.loc[1972, 'ISO code'] = 'FRG'
hosts.loc[1980, 'ISO code'] = 'URS'
hosts.loc[1988, 'ISO code'] = 'KOR'

# Reset Index of hosts: hosts
hosts =hosts.reset_index()

# Print hosts
print(hosts)
```
